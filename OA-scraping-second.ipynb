```python
# import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import urllib3
urllib3.disable_warnings()
```

```python
# get the book's link
npage = 50
links = []

print('Getting book urls')
print()
agent = {
    "User-Agent":'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Safari/537.36',
    'referer':'https://www.google.com/'
}
for i in range(0,npage):
    url = "http://books.toscrape.com/catalogue/category/books_1/page-"+str(i+1)+".html"
    print(url)
    search = requests.get(url, headers=agent)
    
    soup = BeautifulSoup(search.content,'html.parser')
    for link in soup.find_all('div',attrs={'class':'image_container'}):
        link = link.find('a')
        links.append("http://books.toscrape.com/catalogue" + link.get('href')[5:])
    print('.',end='')
    time.sleep(1)
    
    
print()
print('complete')
```

```python
print(links)
```

```python
title = []
rating = []
price = []
availability = []
product_description = []


for book in links:
        # get the content 
        page = requests.get(book)
        soup = BeautifulSoup(page.text)
        # find the title
        try:
            tit = soup.find(class_='col-sm-6 product_main')
            tit = tit.h1.text
            title.append(tit)
        except:
            title.append("")
            pass
        
        # get the price
        try:
            pri = soup.find(class_='col-sm-6 product_main')
            pri = pri.p.text
            price.append(pri[1:])
        except:
            price.append("")
            pass

        # get the rating
        try:
            rat = soup.find(class_='col-sm-6 product_main')
            rat = rat.findAll('p')[2]['class'][1]
            rating.append(rat)
        except:
            rating.append("")
            pass

        # get availability
        try:
            ava = soup.find(class_='col-sm-6 product_main')
            ava = ava.findAll('p')[1].text
            ava = "I"+ava.split("I")[1]
            ava = ava.split(")")[0]+")"
            availability.append(ava)
        except:
            availability.append("")
            pass

        # get product description
        try:
            li = list(i.get_text() for i in soup.find_all('p'))
            pro = max(li, key=len, default='')
            product_description.append(pro)
        except:
            product_description.append("")
            pass
```

```python
from pandas import DataFrame
my_list = list(zip(title,rating,price,availability,product_description))
df = DataFrame(my_list,columns=['Title','Rating','Price','Availability','Description'])
```
